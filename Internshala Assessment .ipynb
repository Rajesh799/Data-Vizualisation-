```python
df1=df.rename(columns={'Activity 1' : 'Activity_1'})
df2 = df1.rename(columns={'Activity 2' : 'Activity_2'})
df2
```

```python
#Replace boolean with proper sentence to understand

df2['Activity_1'] = df2['Activity_1'].map({True: 'Activity1 Detected', False: 'Activity1 Not Detected'})
df2.head(5) 
```

```python
df2['Activity_2'] = df2['Activity_2'].map({True:'Activity2 Detected', False:'Activity2 Not detected'})
df2.head(5)
```

```python
#Removed Date from Timestamp

df2['TimeStamp'] = pd.to_datetime(df['TimeStamp']).dt.time
df2.head(5)
```

```python
#Activities took place from the Date 2022-03-12 to 2022-03-16. Total 5 Days

df2.groupby('Date').groups.keys()
```

```python
# Getting Data only when Activity 1 is Detected

Activity_1 = df2.groupby('Activity_1').get_group('Activity1 Detected')
Activity_1
```
